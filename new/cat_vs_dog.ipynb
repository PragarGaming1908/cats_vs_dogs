```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential, layers
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
t_ds = keras.utils.image_dataset_from_directory(
    directory = "train",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256, 256),
)
v_ds = keras.utils.image_dataset_from_directory(
    directory = "test",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256, 256),
)
```

```python
def process(image, label):
    image = tf.cast(image/255., tf.float32)
    return image,label
t_ds = t_ds.map(process)
v_ds = v_ds.map(process)
```

```python
model = Sequential()

l1_1=Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3))
l2_1=MaxPooling2D(pool_size=(2,2),strides=2,padding='valid')
model.add(l1_1)
model.add(BatchNormalization())
model.add(l2_1)

l1_2=Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu')
l2_2=MaxPooling2D(pool_size=(2,2),strides=2,padding='valid')
model.add(l1_2)
model.add(BatchNormalization())
model.add(l2_2)

l1_3=Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu')
l2_3=MaxPooling2D(pool_size=(2,2),strides=2,padding='valid')
model.add(l1_3)
model.add(BatchNormalization())
model.add(l2_3)

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
opt = keras.optimizers.Adam(learning_rate=0.001, decay=1e-6)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
data = model.fit(t_ds,epochs=10,validation_data=v_ds)
```

```python
model.evaluate(v_ds)
```
